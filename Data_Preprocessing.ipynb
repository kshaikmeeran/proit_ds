## Step3: Data Transformation

### 2 Transformation Techniques: Discrete Vs Continous

**1. Discrete - Label Encoding/One Hot Encoding**

**2. Continous - MinMax Scaler/Standard Scaler/Robust Scaler**

## 1. Use Label Encoding

```python
weather_report_2010_2 = weather_report_2010.copy()
weather_report_2010_2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
weather_report_2010_2['Weather'] = le.fit_transform(weather_report_2010_2['Weather'])
weather_report_2010_2
```

```python
weather_report_2010_2.dtypes
```

## 2. Using OneHot Encoder

It can be built by using 2 libraries:

1. sklearn - OneHotEncoder()
2. Pandas  - pd.get_dummies()

```python
weather_report_2010_3 = weather_report_2010.copy()
weather_report_2010_3
```

#### 2.1 Using Pandas

```python
weather_report_2010_3 = pd.get_dummies(data = weather_report_2010_3,columns=['Weather'])
weather_report_2010_3
```

```python
weather_report_2010_3.dtypes
```

#### 2.2 Using Sklearn

```python
weather_report_2010_4 = weather_report_2010.copy()
weather_report_2010_4
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit_transform(weather_report_2010_4['Weather'])
```

## =====================================================================

## How to choose between OHE and LE????

### Input feature:
Based on whether it is **Parametric or a Non Parametric Model**, we will choosing between LE or OHE.
* For **Parametric Models** - better to go with **OHE.**
* For **Non-Parametric Models** - always go with **Label Encoding.**

### Output feature:
Always the output feature, it must be **Label Encoder.**

```python
def movie(name,n):
    sim=linear_kernel(vectors[indexed[name]],vectors)
    scores=pd.DataFrame(sim).T
    scores.columns=["Score"]
    scores=scores.sort_values("Score",ascending=False)
   
  

    lst=[]
    for i in range(0,n+1):
       
        s={ "title": Movies_df["Movie Title"][scores.index[i]],
           'Ratings':Movies_df["Ratinng"][scores.index[i]],
           'genre':Movies_df["genre"][scores.index[i]]
          }
        lst.append(s)
        
    return lst
    
```
